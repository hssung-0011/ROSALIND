**Enumerating Gene Orders**

A permutation of length n is an ordering of the positive integers {1,2,…,n}. For example, π=(5,3,2,1,4) is a permutation of length 5.

```python
import math
```

```python
number = 6 # input number
```

```python
def add_list(total_length, cur_list):
    return_list = []
    for i in range(total_length):
        if (i+1) not in cur_list:
            return_list.append(cur_list + [i+1])
        else:
            continue
    return return_list
```

```python
def permut(num):
    total_list = add_list(num,[])
    while len(total_list[0]) < num:
        temp_list = []
        save_list = []
        for list in total_list:
            temp_list = add_list(num, list)
            save_list.extend(temp_list)
        total_list = save_list
    return total_list
```

```python
print(math.factorial(number))
result_list = permut(number)
for lists in result_list:
    print(' '.join(map(str,lists)))
```
