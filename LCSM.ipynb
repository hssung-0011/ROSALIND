**Finding a Shared Motif**

 A common substring of a collection of strings is a substring of every member of the collection. We say that a common substring is a longest common substring if there does not exist a longer common substring. For example, "CG" is a common substring of "ACGTACGT" and "AACCGTATA", but it is not as long as possible; in this case, "CGTA" is a longest common substring of "ACGTACGT" and "AACCGTATA".


```python
filepath = "/mnt/c/Data/ROSALIND_download/rosalind_lcsm.txt"
seq_dict = {}
with open(filepath) as f:
    for line in f.readlines():
        line = line.rstrip()
        if line.startswith(">"):
            id_temp = line
            seq_dict[id_temp] = ""
        else:
            seq_dict[id_temp] += line
            
seq_list = list(seq_dict.values())
seq_list
```

first_common 함수 정의 -> seq, num을 입력으로 받는다.
seq를 구성하는는 num 길이의 모든 모티프를 리스트로 반환한다.

```python
def first_common(num, seq):
    output_list = [seq[i:i+num] for i in range(len(seq)+1 - num) if (seq[i:i+num] in seq_list[1])]
    return output_list
```

common motif의 후보가 될 수 있는 모든 motif를 common list에 저장한다.
seq의 전체 길이부터 2까지 순서대로 first common  돌려서 common list에 저장

```python
common_list = []
for i in range(len(seq_list[0]),1,-1):
    common_list += first_common(i, seq_list[0])
# common_list
```

common_find 정의. -> common list와 seq를 입력으로 받는다.
common list의 모든 모티프 중 seq에 존재하는 모티프만 필터링.

```python
def common_find(common_list, seq):
    common_list = [i for i in common_list if i in seq]
    return common_list
```

모든 서열에 대해 common_find 돌린다.
이 때 모든 서열에서 나타나는 common find만 남게 된다. -> 남은 common find motif 중 첫번째로 나오는 값 프린트.

```python
for seq in seq_list[2:]:
    common_list = common_find(common_list, seq)

print(common_list[0])
```
