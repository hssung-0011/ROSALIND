**Genome Assembly as Shortest Superstring**

For a collection of strings, a larger string containing every one of the smaller strings as a substring is called a superstring.

By the assumption of parsimony, a shortest possible superstring over a collection of reads serves as a candidate chromosome.

```python
from Bio import SeqIO
```

```python
filepath = "/mnt/c/Data/ROSALIND_download/rosalind_long.txt"
sequences = []
for seqs in SeqIO.parse(filepath, "fasta"):
    print(seqs.seq)
    sequences.append(str(seqs.seq))
    
```

두 서열 비교해서 겹치는 길이 반환

```python
def compare(seq1, seq2):
    """seq1을 머리 seq2를 꼬리라 가정. 이 때 두 서열의 겹치는 길이 반환"""
    for i in range(len(seq1)):
        if seq2.startswith(seq1[i:]):
            return len(seq1[i:])
    return 0
```

**각 서열에 대해 '겹치는 서열이 가장 긴 것의 길이'와 '가장 많이 겹치는 서열의 위치'를 저장**

```python
l = len(sequences)
m = [0]*l # 겹치는 서열 중 가장 긴 것의 길이
p = [-1]*l # 겹치는 서열의 위치

for i in range(l):
    for j in range(l):
        if j == i:
            continue
        
        temp = compare(sequences[i], sequences[j])
        if temp > m[i]:
            m[i] = temp
            p[i] = j

# 겹치는 길이가 가장 적은 것 = 끝 서열
p[m.index(min(m))] = -1
print(p, m)
p_dict = {i:p.index(i) for i in p}

print(sorted(p))
        

        
```

```python

def assemb(seq1, seq2):
    '''seq1을 머리, seq2를 꼬리로 두고 두 서열 합치기. 그중 가장 짧은 서열 반환'''
    for i in range(len(seq1)):
        if seq2.startswith(seq1[i:]):
            return seq1[:i] + seq2  
```

**끝 서열을 시작으로 모든 서열 합치기**

```python

cur = -1
complete_seq = sequences[p.index(cur)]
while p.index(cur) in p:
    cur = p.index(cur)
    seq_a = sequences[p.index(cur)]
    complete_seq = assemb(seq_a, complete_seq)
    
    
```
